## 1.1 Regression

**<font color='blue'>Regression algorithms </font>** is a type of supervised learning algorithm learns to predict numbers out of infinitely many possible numbers

Let's say we want to predict housing prices based on the size of the house. We have collected some data and plot the data and it looks like this.



With this data, let's say a friend wants to know what's the price for their 750 square foot house. How can the learning algorithm help you?

<font color='red'>Answer: </font> In a regression application like predicting prices of houses, the learning algorithm has to predict numbers from infinitely many possible output numbers. From the plot, we have information that a 750 square foot house could be sold for maybe about $\$150,000$. Or our friend's house could be sold for closer to $200,000.


The task of the learning algorithm is to produce more of these right answers, specifically predicting what is the likely price for other houses like our friend's house. That's why this is supervised learning. To define a little bit more terminology, this housing price prediction is the particular type of <font color='blue'>supervised learning </font> called <font color='blue'>regression </font>.


By regression, we are trying to predict a number from infinitely many possible numbers such as the house prices in our example, which could be $150,000$ or $70,000$ or $183,000$ or any other number in between.

**That's <font color='blue'>supervised learning </font>, learning input, output, or x to y mappings. You saw in this video an example of regression where <font color='blue'>the task is to predict number</font>.**



## 1.2 Classification

**<font color='blue'>In classification the learning algorithm has to make a prediction of a category, all of a small set of possible outputs.</font>**

Let's take **breast cancer detection** as an **<font color='blue'>example</font>**.

Say you're building a machine learning system so that doctors can have a diagnostic tool to detect breast cancer. This is important because early detection could potentially save a patient's life. Using a patient's medical records your machine learning system tries to figure out if a tumor that is a lump is **<font color='green'> malignant meaning cancerous or dangerous</font>**. Or if that tumor, that lump is **<font color='green'>benign, meaning that it's just a lump that isn't cancerous and isn't that dangerous</font>**?

These tumors are labeled as either **<font color='green'>benign</font>**, which we will designate in this example with a **0** or **<font color='green'>malignant</font>** , which will designate in this example with a **1**.

**<font color='red'>One reason that this is different from regression is that we're trying to predict only a small number of possible outputs or categories. In this case two possible outputs 0 or 1, benign or malignant.</font>** So the fact that there are only two possible outputs is what makes this classification.


Let's call two different types of cancer type 1 and type 2. In this case the average would have **three possible output categories** it could predict **<font color='green'>(benign, malignant type 1, and malignant type 2).</font>**
